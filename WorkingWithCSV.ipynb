## **6. usecols parameter**

## **7. Skiprows/nrows parameter**

## **8. Encoding parameter**

## **9. Skip bad lines**

```python
pd.read_csv('/content/sample_data/california_housing_test.csv',skiprows=[1,2], nrows = 100, encoding='utf-8',on_bad_lines='skip') # warn, skip, error by default
```

## **10. dtypes parameter**

```python
pd.read_csv('/content/sample_data/california_housing_test.csv',dtype={'housing_median_age':int}).info()
```

## **11. Handling Dates**

```python
pd.read_csv('/content/sample_data/california_housing_test.csv', parse_dates=[])
```

## **12. Convtors**

```python
def change(latitude):
    if latitude == 0.0:
        return 0
    else:
        return latitude

pd.read_csv(
    '/content/sample_data/california_housing_test.csv',
    converters={'latitude': change}
)
```

## **13. na_values parameter**

```python
pd.read_csv('/content/sample_data/california_housing_test.csv',na_values=['33.78'])
```

## **14. Loading a huge dataset in chunks**

```python
df= pd.read_csv('/content/sample_data/california_housing_test.csv',chunksize = 300)
```
