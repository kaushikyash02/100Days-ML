```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import PromptTemplate
```

```python
books = {}

for doc in docs:
    title = doc.metadata.get("title")
    if title not in books:
        books[title] = doc.metadata["total_pages"]

print(books)
```

```python
book = [
    doc for doc in docs
    if doc.metadata["title"] == "Nineteen Eighty-Four"
]
```

```python
print(book[0].metadata["author"])
print(book[0].metadata["total_pages"])
print(book[0].metadata["title"])
```

```python
book_text = "\n".join(doc.page_content for doc in book)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    google_api_key="API_KEY",
    temperature=0.7
)
prompt = PromptTemplate(
    template="""
how many pages does Nineteen Eighty-Four have?

{text}
""",
    input_variables=["text"]
)
parser = StrOutputParser()

chain = prompt | llm | parser
```

```python
summary = chain.invoke({"text": book_text})

print(summary)
```
